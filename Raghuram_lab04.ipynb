```python
#computing the Z-score for Age and Fare.

#accessing the coulumn data..
Age_data = Updated_data['Age'].tolist()
Fare_data = Updated_data['Fare'].tolist()


#mean

mean_value_Age = np.mean(Age_data)
mean_value_Fare = np.mean(Fare_data)

#standard deveation

std_value_Age = np.std(Age_data)
std_value_Fare = np.std(Fare_data)

#print(mean_value_Age,mean_value_Fare,std_value_Age,std_value_Fare)


#empty list to append the data.

New_Age_data,New_Fare_data = [],[]

for i in range(len(Age_data)):
    New_Age_data.append((Age_data[i] - mean_value_Age)/std_value_Age)
    New_Fare_data.append((Fare_data[i] - mean_value_Fare)/std_value_Fare)

#print(len(New_Age_data))

#updating the Age,Fare columns with Z-score

Updated_data['Age'] = New_Age_data
Updated_data['Fare'] = New_Fare_data


#print(New_Age_data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, zero_one_loss

#  excluding PassengerId and Survived as i feel they don't need for training
X = Updated_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
#  target variable
y = Updated_data['Survived']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

# Determine the best K value 
best_k = None
best_accuracy = 0
k_values = range(1,30)
accuracy_values = []
error_rates = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_values.append(accuracy)
    
    error = zero_one_loss(y_test, y_pred)
    error_rates.append(error)
    
    if accuracy > best_accuracy:
        best_k = k
        best_accuracy = accuracy

#the best K value and its corresponding accuracy
print(f"The best K value is {best_k} with an accuracy of {best_accuracy}")

# the accuracy vs. K value graph
plt.plot(k_values, accuracy_values)
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. K Value')
plt.show()


new_passenger_features = test_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
new_prediction = knn.predict(new_passenger_features)

count = 0
for i in range(len(new_prediction)):
    if new_prediction[i] == 1:
        count+=1

print(f"Number of predicted survivors: {count}")
```
